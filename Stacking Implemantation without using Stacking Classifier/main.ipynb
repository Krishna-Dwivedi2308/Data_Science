import the required libraries first

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

Load the dataset 

```python
data=pd.read_csv('.\diabetes.csv')
data.head()
```

Split the data into input and target features 

```python
X=data.drop('Outcome',axis=1)
y=data['Outcome']

print(X)
print(y)
```

split the data into two subsets :
- Subset 1(Train Sets)
- Subset 2(Validation Set/Hold Out Set)

```python
from sklearn.model_selection import train_test_split
train,val_train,test,val_test=train_test_split(X,y,test_size=0.5,random_state=44)
```

Split the train set further into train and test set 

```python
X_train,X_test,y_train,y_test=train_test_split(train,test,test_size=0.2,random_state=44)
```

### Model 1-Implemantation of KNeighbour Classifier 

```python
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)
y_pred_knn
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_knn)
```

### Model 2- Implementation of SVC Classifier

```python
svc=SVC()
svc.fit(X_train,y_train)
y_pred_svc=svc.predict(X_test)
y_pred_svc  
```

```python
accuracy_score(y_test,y_pred_svc)
```

Predictions of the model in the validation set 

```python
prediction_val1=knn.predict(val_train)
prediction_val2=svc.predict(val_train)

print('KNN Prediction\n',prediction_val1)
print('svc Prediction\n',prediction_val2)
```

```python
predict_val=np.column_stack((prediction_val1,prediction_val2))
predict_val.shape
```

```python
predict_test1=knn.predict(X_test)
predict_test2=svc.predict(X_test)
predict_test=np.column_stack((predict_test1,predict_test2))
predict_test
```

### Input featurs are the predictions of the model M1(KNN) and model M2(SVC) stored inside the predict_val (input features) for the Random Forest Classifier 

```python
print('Predict_val\n',predict_val.shape)
print('Val_test \n',val_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rand_clf=RandomForestClassifier()
rand_clf.fit(predict_val,val_test)
```

```python
rand_clf.score(predict_test,y_test)
```

**Hyperparameter Tuning**

```python
grid_param={
"n_estimators":[100,300,500],
"max_features":["auto",'log2'],
"criterion":["gini","entropy"],
"min_samples_split":[4,5,6,7,8],
"min_samples_leaf":[1,2,3,4,5 ]
}

from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=rand_clf,param_grid=grid_param,cv=5,n_jobs=-1,verbose=3 )
```

```python
grid_search.fit(predict_val,val_test)
print(grid_search.best_params_)
```

- These come out to be the best parameters : **{'criterion': 'gini', 'max_features': 'log2', 'min_samples_leaf': 1, 'min_samples_split': 4, 'n_estimators': 100}**
