```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
# following package to create dendogram 
from scipy.cluster.hierarchy import dendrogram, linkage
```

Load the data 

```python
X, y = load_iris(return_X_y=True)
```

Dendrogram creation


```python
Z = linkage(X, method='ward')
plt.figure(figsize=(20, 10))
plt.title('Iris Data Dendrogram')
dendrogram(Z)
plt.show()
```

The Blue line is where the clusters are getting matured . Cut through that line to get the clusters required . Here = 3.(see notes)

```python
model=AgglomerativeClustering(n_clusters=3)
labels=model.fit_predict(X)
```

Visualize the Clusters
Using the first two features    

```python
plt.figure(figsize=(20,10))
plt.scatter(X[:,0],X[:,1],c=labels)
# we can check what the first two features are 
plt.xlabel("Sepal Langth")
plt.ylabel("Sepal Width")
plt.title("Agglomerative Clustering")
plt.show()
```
